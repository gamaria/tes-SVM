```python
import pandas as pd
import csv
import re
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.cross_validation import train_test_split
#from sklearn.naive_bayes import MultinomialNB
from sklearn.feature_extraction.text import TfidfVectorizer
```

# Input Data

```python
url = open ("dataTanya.csv", "r")
names =['pertanyaan','kelas'] #disi harus mucul hasil dari set token
dataTanya = pd.read_csv(url, names=names)
dataTanya
```

```python
len(dataTanya)
```

```python
tanya_x=dataTanya["pertanyaan"]
kelas_y=dataTanya["kelas"]
print(dataTanya.groupby("kelas").size())
```

## Inisialiasi Dataset

```python
validation_size = 0.30
seed = 2
x_train, x_test, y_train, y_test = train_test_split(tanya_x, kelas_y, test_size=validation_size, random_state=seed)
```

```python
#cek dataseet
x_train
```

#keterangan
x_train: dataset atribut yang akan ditraning,
y_train: dataset kelas yang akan di dtraning,
x_test : dataset atribut yang akan ditesting,
y_test : dataset kelas yang akan ditesting.

## Tokenisasi 

```python
def wm2df(wm, feat_names):
    
    # create an index for each row
    doc_names = ['Doc{:d}'.format(idx) for idx, _ in enumerate(wm)]
    df = pd.DataFrame(data=wm.toarray(), index=doc_names,
                      columns=feat_names)
    return(df)

cvec = CountVectorizer(lowercase=True)
wm = cvec.fit_transform(x_train)
tokens = cvec.get_feature_names()
wm2df(wm, tokens)
```

```python
#tokrn ---> untuk cek token yang dihasilkan
totalToken= len (tokens)
totalToken
```

## Vektorisasi TF-IDF

```python
#TRANING TF-IDF
cv = TfidfVectorizer(min_df=1)
w_tanya=cv.fit_transform(x_train)
w_tanya=w_tanya.toarray()
#w_tanya
```

```python
#cv.get_feature_names()
```

```python
#TESTING TF_IDF
x_testcv = cv.fit_transform(x_test)
x_testcv = x_testcv.toarray()
x_testcv
```

```python
#inisialisasi dataset untuk ditraning dalam SVM
u= totalToken
dataSVM= w_tanya
dataKelas = y_train.values
data_atribut = dataSVM[:,0:u]
scoring = 'accuracy'
```

```python
dataKelas
```

```python
from sklearn import svm
from sklearn import model_selection
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score


svm_model = svm.SVC(probability=True)
svm_model.fit(data_atribut,dataKelas )
predictions = svm_model.predict(x_testcv)
print('Accuracy =', accuracy_score(y_test, predictions))
print(confusion_matrix(y_test, predictions))
print(classification_report(y_test, predictions))
```

```python
#CEK NILAI KEBENARAN
prediksi_benar =(predictions == y_test).sum()
prediksi_salah = (predictions != y_test).sum()
cekprediksi= prediksi_benar/(prediksi_benar+prediksi_salah)
print('predikis benar adalah',prediksi_benar)
print('prediksi_salah', prediksi_salah)
print('presentasi', cekprediksi)
```
